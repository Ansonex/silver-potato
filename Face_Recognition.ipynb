```python
import cv2
import face_recognition
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/messi.jpg')
```

```python
plt.imshow(img)
```

```python
imgMessi = cv2.imread('/content/drive/MyDrive/messi.jpg',cv2.IMREAD_UNCHANGED)
imgMessi = cv2.cvtColor(imgMessi,cv2.COLOR_BGR2RGB)
imgMessiTest = cv2.imread('/content/drive/MyDrive/imgMessiTest.png',cv2.IMREAD_UNCHANGED)
# imgMessiTest = cv2.cvtColor(imgMessiTest,cv2.COLOR_BGR2RGB)

cv2_imshow(imgMessi)
cv2_imshow(imgMessiTest)
```

```python
faceLoc = face_recognition.face_locations(imgMessi)[0]
encodeMessi = face_recognition.face_encodings(imgMessi)[0]
print(faceLoc)
```

(44, 164, 95, 112) top, right,bottom,left - give the x1y1 and x2y2

```python
cv2.rectangle(imgMessi,(faceLoc[3],faceLoc[0]),(faceLoc[1],faceLoc[2]),(255,0,255),2)
```

```python
cv2_imshow(imgMessi)
```

```python
faceLoc2 = face_recognition.face_locations(imgMessiTest)[0]
encodeMessi2 = face_recognition.face_encodings(imgMessiTest)[0]
print(faceLoc2)
```

(118, 563, 341, 340) top, right,bottom,left - give the x1y1 and x2y2

```python
cv2.rectangle(imgMessiTest,(faceLoc2[3],faceLoc2[0]),(faceLoc2[1],faceLoc2[2]),(255,0,255),2)
```

```python
cv2_imshow(imgMessiTest)
```

## Compare the face

```python
results = face_recognition.compare_faces([encodeMessi],encodeMessi2)
print(results)
```

```python
facedis = face_recognition.face_distance([encodeMessi],encodeMessi2)
print(facedis)
```

value more low mean distance more close

```python
cv2.putText(imgMessiTest,f'{results} {round(facedis[0],2)}',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
cv2_imshow(imgMessiTest)
```
